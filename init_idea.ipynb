```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")

inputs = tokenizer("A step by step recipe to make bolognese pasta:", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

```python
# Define the quiz generation prompt and context documents
prompt = "Generate a quiz question about photosynthesis."
document_1 = "Photosynthesis is the process by which plants use sunlight, water, and carbon dioxide to create oxygen and energy in the form of sugar."
document_2 = "The process of photosynthesis occurs in the chloroplasts of plant cells and is essential for producing the oxygen that most organisms need to survive."

# Combine the prompt and documents into a structured input
input_text = f"Instruction: {prompt} Context: {document_1} Additional Context: {document_2}"
```

```python
# Encode the input and generate the output
input_ids = tokenizer.encode(input_text, return_tensors="pt")
outputs = model.generate(input_ids, max_length=50, num_beams=5, early_stopping=True)

# Decode the generated output to obtain the quiz question
quiz_question = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Quiz Question:", quiz_question)
```

```python
import fitz  # PyMuPDF

# Load the PDF
pdf_path = "/content/L4c- lists.pdf"
doc = fitz.open(pdf_path)

# Extract text from each page
text_content = []
for page in doc:
    text = page.get_text("text")
    text_content.append(text)

# Combine all extracted text
full_text = " ".join(text_content)
```

```python
import nltk
nltk.download('punkt')
```

```python
import nltk
from nltk.tokenize import sent_tokenize

# Split the full text into sentences
sentences = sent_tokenize(full_text)
```

```python
sentences
```

```python
# Define prompt and relevant document content for a section
prompt = "Generate a quiz question. and it possible answers"
document_content = " ".join(sentences)  # Use a subset of sentences relevant to "functions" if possible

# Prepare the input for FLAN-T5 or any other model
input_text = f"Instruction: {prompt} Context: {document_content}"
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load FLAN-T5 model and tokenizer
model_name = "google/flan-t5-base"
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = T5Tokenizer.from_pretrained(model_name)
```

```python


# Encode and generate questions
input_ids = tokenizer.encode(input_text, return_tensors="pt")
outputs = model.generate(input_ids, max_length=50, num_beams=5, early_stopping=True)

# Decode and print the quiz question
quiz_question = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Quiz Question:", quiz_question)
```
